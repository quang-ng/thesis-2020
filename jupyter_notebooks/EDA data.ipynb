```python
import matplotlib.pyplot as plt
```

```python
battery_v=[3.54,3.64,3.71,3.73,3.78,3.84,3.92,4.05,4.14,4.07,4.05]
time=[0,12,24,36,48,60,72,84,96,108,120]

plt.plot(time, battery_v, linestyle="-", color='black')
plt.xlabel('Time(min)', fontsize=14)
plt.ylabel('Voltage (V)', fontsize=14)
plt.show()
```

```python
battery_current=[1.30,1.30, 1.30, 1.30, 1.30, 1.30, 1.30, 0.0, 0.0, 0.0, 0.0]
time=[0,12,24,36,48,60,72,84,96,108,120]

plt.plot(time, battery_current, linestyle="-", color='black')
plt.xlabel('Time(min)', fontsize=14)
plt.ylabel('Current(A)', fontsize=14)
plt.show()
```
